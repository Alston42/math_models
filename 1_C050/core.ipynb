```python
## 灰色关联分析

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Load gdp.mat
# Read data from excel file
data = np.zeros((12, 4, 6))
data[:, :, 0] = np.array(pd.read_excel('六大蔬菜品类与其节日、节气和季度蔬菜丰富度指标.xlsx', sheet_name='花菜类'))
data[:, :, 1] = np.array(pd.read_excel('六大蔬菜品类与其节日、节气和季度蔬菜丰富度指标.xlsx', sheet_name='食用菌'))
data[:, :, 2] = np.array(pd.read_excel('六大蔬菜品类与其节日、节气和季度蔬菜丰富度指标.xlsx', sheet_name='花叶类'))
data[:, :, 3] = np.array(pd.read_excel('六大蔬菜品类与其节日、节气和季度蔬菜丰富度指标.xlsx', sheet_name='辣椒类'))
data[:, :, 4] = np.array(pd.read_excel('六大蔬菜品类与其节日、节气和季度蔬菜丰富度指标.xlsx', sheet_name='茄类'))
data[:, :, 5] = np.array(pd.read_excel('六大蔬菜品类与其节日、节气和季度蔬菜丰富度指标.xlsx', sheet_name='水生根茎类'))

for i in range(6):
    Mean = np.mean(data[:, :, i])
    data[:, :, i] = data[:, :, i] / np.tile(Mean, (data[:, :, i].shape[0], 1))
    Y = data[:, 0, i]
    X = data[:, 1:, i]
    absX0_Xi = np.abs(X - np.tile(Y, (1, X.shape[1])))
    a = np.min(np.min(absX0_Xi))
    b = np.max(np.max(absX0_Xi))
    rho = 0.5
    gamma = (a + rho * b) / (absX0_Xi + rho * b)
    print("子序列中各个指标的灰色关联度分别为：")
    print(np.mean(gamma))
    x = np.arange(1, 13)
    plt.figure(i)
    plt.plot(x, data[:, 0, i], 'k', linewidth=1)
    plt.plot(x, data[:, 1, i], 'r', linewidth=1)
    plt.plot(x, data[:, 2, i], 'b', linewidth=1)
    plt.plot(x, data[:, 3, i], 'g', linewidth=1)
    plt.axis('on')
    plt.box('on')

plt.show()
```
